```python
# Mode for Categorical 
df['Product']=df['Product'].fillna(df['Product'].mode()[0])
df['TypeName']=df['TypeName'].fillna(df['TypeName'].mode()[0])
df['OS']=df['OS'].fillna(df['OS'].mode()[0])
df['Screen']=df['Screen'].fillna(df['Screen'].mode()[0])
```

```python
# Mean and Median for Numerical
df['Price_euros']=df['Price_euros'].fillna(df['Price_euros'].mean())
df['CPU_freq']=df['CPU_freq'].fillna(df['CPU_freq'].median())
```

```python
# After Handling the missing values:
df.isnull().sum()
```

```python
# Handling Duplicates
df.duplicated().sum()
df=df.drop_duplicates()
```

```python
# Detect and treat outliers using IQR method 
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Keep only rows within bounds
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Example usage for multiple columns
df_clean = remove_outliers_iqr(df, ['Price_euros', 'CPU_freq'])

print(df_clean)

```

```python
# Z-score
import numpy as np
from scipy import stats

def remove_outliers_zscore(df, column, threshold=3):
    z_scores = stats.zscore(df[column], nan_policy='omit')
    return df[(abs(z_scores) < threshold)]

df_zscore=remove_outliers_zscore(df,'Price_euros')
```

```python
# Feature selection using Variance Threshold
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.6)  # Remove features with variance < 0.2
X_reduced = selector.fit_transform(df)

# Selected feature names
selected = df.columns[selector.get_support()]
print("Features kept after Variance Threshold:", list(selected))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Select only numerical features
X = df.select_dtypes(include=['int64', 'float64'])

# Apply PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create DataFrame for visualization
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])

# Plot PCA results
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.6)
plt.title('PCA - 2 Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import IsolationForest
x_col = "Price_euros"
y_col = "Weight"
df_vis = df[[x_col, y_col]].dropna()
iso = IsolationForest(contamination=0.05, random_state=42)
y_pred = iso.fit_predict(df_vis)

plt.figure(figsize=(8,6))
plt.scatter(
    df_vis[x_col], df_vis[y_col],
    c=['red' if p == -1 else 'blue' for p in y_pred],
    alpha=0.6
)

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title("Isolation Forest Outlier Detection")
plt.legend(["Inliers", "Outliers"], loc="best")
plt.show()
```
