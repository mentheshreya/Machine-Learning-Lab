```python
df = df.drop(["baby_id", "name", "date"], axis=1)
```

```python
print(df.isnull().sum())
```

```python
# Drop 'apgar_score' due to too many missing values
df = df.drop(columns=["apgar_score"])
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le_risk = LabelEncoder()
cat_cols = df.select_dtypes(include=['object']).columns.drop("risk_level")
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Encode the target variable 'risk_level'
df["risk_level"] = le_risk.fit_transform(df["risk_level"])
```

```python
leaky_features = ['jaundice_level_mg_dl', 'oxygen_saturation', 'heart_rate_bpm']
X = df.drop(columns=["risk_level"] + leaky_features)
y = df["risk_level"]
```

```python
print(" After Encoding ")
print(df.head())
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, RocCurveDisplay, precision_score, recall_score, f1_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print("\nTraining Samples:", X_train.shape[0])
print("Testing Samples:", X_test.shape[0])
```

```python
# Using Entropy
dt = DecisionTreeClassifier(
    criterion="entropy",  # use information gain
    max_depth=5,          # limit depth
    random_state=42
)

# Train the model
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\n Accuracy:", accuracy)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f"Precision : {precision:.4f}")
print(f"Recall : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt, feature_names=X.columns, class_names=[str(c) for c in le.classes_],
          filled=True, rounded=True, fontsize=12)
plt.title("Decision Tree for Newborn Risk Prediction")
plt.show()
```

```python
# Feature Importance
importances = pd.Series(dt.feature_importances_, index=X.columns)
print("\n===== Top Features =====")
print(importances.sort_values(ascending=False).head(10))
```

```python
# SELF LEARNING
from sklearn.metrics import roc_curve, roc_auc_score, RocCurveDisplay

# Make probability predictions for the positive class
y_prob = dt.predict_proba(X_test)[:, 1]  # probability of "At Risk"

# Compute ROC AUC score
auc_score = roc_auc_score(y_test, y_prob)
print("\n ROC-AUC Score:", auc_score)

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color="blue", label=f"ROC curve (AUC = {auc_score:.2f})")
plt.plot([0,1], [0,1], color="gray", linestyle="--")  # diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for Newborn Risk Prediction")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

```
